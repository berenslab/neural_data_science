```python
import pandas as pd
import numpy as np
import os

from morphopy.neurontree.plotting import show_threeview
from morphopy.neurontree import NeuronTree as nt

import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set_context('notebook')
```

```python
def load_swc(filepath):
    """ 
    Loads in the swc located at filepath as a pandas dataframe.
    """
    swc = pd.read_csv(filepath, delim_whitespace=True, comment='#',
                      names=['n', 'type', 'x', 'y', 'z', 'radius', 'parent'], index_col=False)
    # convert back into microns as the data was divided by 100 in x and y
    swc[['x', 'y', 'z']] *= 100
    return swc

def load_morphology(filepath):
    """ Loads in the swc located at filepath and creates and returns a NeuronTree object. """
    swc = load_swc(filepath)
    N = nt.NeuronTree(swc=swc)
    
    return N

def load_files(path):
    """ Returns an object array of NeuronTrees containing all reconstructions at `path`. 
    The reconstructions are sorted ascendingly by their filename."""
    neurons = []
    path, _, files = list(os.walk(path))[0]
    files.sort()
    for f in files:
        N = load_morphology(path+f)
        neurons.append(N)
    return neurons

colors = sns.color_palette('rainbow_r', n_colors=14)
```

#### File format

Morphological reconstructions are typically stored in the SWC file format, a simple text file that holds node information in each row and connects nodes through the `parent` node id. A parent id of -1 indicates no parent, so the starting point of the tree graph. 
The `type` label indicates the node type (1: somatic , 2: axonal, 3: dendritic (basal), 4: dendritic (apical), 5+: custom).
The code snippet below loads in one swc file and prints its head. 

You can find a more detailed specification of SWC and SWC+ [here](http://www.neuronland.org/NLMorphologyConverter/MorphologyFormats/SWC/Spec.html) and [here](https://neuroinformatics.nl/swcPlus/).

SWC files are a compact way for storing neural morphologies but their graph structure makes them difficult to handle for current machine learning methods. We, therefore, need to convert our reconstructions into a reasonable vector-like representations. 

Here we will present two commonly chosen representations: Morphometric statistics and density maps


## Morphometric statistics

Morphometric statistics denote a set of hand-crafted single valued features such as `soma radius`, `number of tips` or `average branch angle`. For a more detailed explanation of morphometrics please refer to the [MorphoPy documentation](https://github.com/berenslab/MorphoPy#morphometric-statistics).

First we want to compute this representation for each cell. For this, we can use the function `compute_morphometric_statistics` of the MorphoPy package which computes a predefined set of $28$ statistics.

```python
from morphopy.computation.feature_presentation import compute_morphometric_statistics
# extraction of morphometric statistics on entire data set
ms_list = [compute_morphometric_statistics(n) for n in neurons] # this line might take a while

# concatenate data into one table and set the Cell_nr as index
morphometric_statistics = pd.concat(ms_list).set_index(labels['Cell_nr'])

morphometric_statistics.head()
```

Now we want to visualize the data, so go ahead and plot all statistics for each cluster. 

```python
# visualize. Which are the most discriminative ones?

features = morphometric_statistics.columns.values

fig, axes = plt.subplots(4,7, figsize=(25,10), sharex = True)
axes = axes.flatten()
for k, feature in enumerate(features):
    
    for l in np.unique(labels['cluster']):
       
        label_index = (labels['cluster'] == l).values

        axes[k].scatter([l]*sum(label_index), morphometric_statistics[label_index][feature], c=[colors[l-1]])
    axes[k].set_title(feature.replace("_", " "))
axes[-1].set_xticks(range(1,15))
axes[-1].set_xlabel('Cluster id')
sns.despine()
```

**QX.a:** Which statistics separate clusters well? Which can be removed?

**Answer:** First, there is no single statistic that separates all clusters reliably. However, we see differences for e.g. `total length`, `height`, the number of `branch points` and `mean branch angle`.

Second, all thickness related statistics need to be removed since the radius was not traced (--> `avg thickness`, `max thickness`, `total surface` and `total volume`).
Additionally, we need to remove ` log min tortuosity` as it carries no signal either.



**QX.b:** More generally, what do morphometric statistics capture well? What are their advantages, what might be their downsides? Briefly explain.

**Answer:** Morphometric statistics are easy to extract and amneable to classical ML application. Obvious differences in size or branching can be captured through them. However, morphometrics abstract away a lot of the fine detail of a morphologie's geometry. Also several different morphologies might map to the same set of morphometrics. 

```python
# For further analysis we will remove uninformative features and z-score along each statistic
features = ['branch_points', 'width', 'depth', 'height', 'tips', 'stems',
       'total_length', 'max_path_dist_to_soma', 'max_branch_order',
       'max_path_angle', 'min_path_angle', 'median_path_angle',
       'mean_soma_exit_angle', 'max_segment_path_length',
       'median_intermediate_segment_pl', 'median_terminal_segment_pl',
       'log_max_tortuosity', 'log_median_tortuosity',
       'max_branch_angle', 'min_branch_angle', 'mean_branch_angle',
       'max_degree', 'tree_asymmetry']

# z-score morphometrics and remove nans and uninformative features
morphometric_data = (morphometric_statistics - morphometric_statistics.mean())/morphometric_statistics.std()
# morphometric_data = morphometric_statistics
morphometric_data[morphometric_data.isna()] = 0
morphometric_data = morphometric_data[features].values
```

## Density maps

Density maps project a neuron's 3D point cloud ($x$, $y$, $z$) onto a plane or an axis, and bin the projected point cloud into a fixed number of bins. Hereby, the binning controls how much global or local information is kept, which majorly affects the results.
While density maps ignore the connectivity between nodes and only assign value to neurite density, this representation can be powerful, when cell type and location in the brain area are highly correlated.

Make sure that you normalize the density maps globally and bin each direction into $20$ bins.

```python
# Find the minimal and maximal x, y,z - coordinates in the data set to normalize the density maps globally
minimal_coordinates = np.vstack([np.percentile(list(n.get_node_attributes('pos').values()), 5,axis=0) for n in neurons])
maximal_coordinates = np.vstack([np.percentile(list(n.get_node_attributes('pos').values()),95, axis=0) for n in neurons])

print('Minimal coordinates: ', minimal_coordinates.min(axis=0).round())
print('Maximal coordinates: ', maximal_coordinates.max(axis=0).round())
```

```python
from morphopy.computation.feature_presentation import compute_morphometric_statistics
config_global = dict(distance=1, n_bins_x=20, n_bins_y=20, n_bins_z=20, 
              density=True,smooth=True, sigma=1, 
               r_max_x=400, r_max_y=450, r_max_z=120, 
               r_min_x=-550, r_min_y=-390, r_min_z=-100)

density_maps= [compute_density_maps(n, config_params=config_global) for n in neurons] # this line might take a while

# extract the z density map
# DM_Z = [550, 20]
DM_Z = np.array([dm['z_proj']['data'] for dm in density_maps])
```

```python
# plot the Z-density map sorted by class label

fig, axes = plt.subplots(1, 14, figsize=(21,3), sharey=True)
axes = axes.flatten()
for l in np.unique(labels['cluster']):
    
    label_index = (labels['cluster'] == l).values
    axes[l-1].plot(DM_Z[label_index].T, range(20), c='lightgrey')
    axes[l-1].plot(np.mean(DM_Z[label_index], axis=0), range(20), c=colors[l-1], linewidth=5)
    axes[l-1].set_title('Cluster %i'%l)
    
sns.despine()
axes[0].set_yticks([0,5,10,15,19])
axes[0].set_yticklabels(density_maps[0]['z_proj']['edges'][0][[0,5,10,15,19]].astype(int))
_ = axes[0].set_ylabel('Z [microns]')
```

```python
from sklearn.decomposition import PCA
pca = PCA()

# PCA on density maps
DM_pca = pca.fit_transform(DM_Z)

# plot scatter plots of the first three principal components
fig, axes = plt.subplots(1,3,figsize=(12,4))
for l in np.unique(labels['cluster']):
    label_index = (labels['cluster'] == l).values
        
    axes[0].scatter(DM_pca[label_index][:,0], DM_pca[label_index][:,1], c=[colors[l-1]])
    axes[1].scatter(DM_pca[label_index][:,0], DM_pca[label_index][:,2], c=[colors[l-1]])
    axes[2].scatter(DM_pca[label_index][:,1], DM_pca[label_index][:,2], c=[colors[l-1]])
sns.despine()
axes[0].set_xlabel('PC1')
axes[0].set_ylabel('PC2')

axes[1].set_xlabel('PC1')
axes[1].set_ylabel('PC3')

axes[2].set_xlabel('PC2')
axes[2].set_ylabel('PC3')
```

Typically, most of the data variance is captured by only a few principal components. Reduce the dimensionality of the density maps such that more than 95% of variance are kept.

```python
n_components = np.where(np.cumsum(pca.explained_variance_ratio_) > .95)[0][0] + 1
print(n_components, 'principal components capture more than 95% of the variance explained.')

DM_pca = DM_pca[:,:n_components]
```

## Hierarchical clustering

Typically, we do not have access to ground truth cell type labels but we need to infer the number of classes that are present in the data. In the following we want to explore hierarchical clustering based on morphometric statistics and on density maps. Assume that you know the number of existing cluster ($N_{clus}=14$). Visualize the clustering dendrogram (use seaborn's `clustermap` method) and then cluster the data and assess the cluster label correspondence with ground truth labels using XXX.

### Clustering on morphometric statistics

```python
true_labels = labels['cluster'].values
sns.clustermap(morphometric_data, method='ward',
               row_colors=[colors[t-1] for t in true_labels],
               cbar_pos=(1., 0.6, 0.05, 0.18), 
               xticklabels = [f.replace("_", " ").capitalize() for f in features])
```

```python
# Do sth with clustering?

from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import mutual_info_score, adjusted_rand_score

agglom = AgglomerativeClustering(n_clusters=14, linkage='ward')
agglom.fit(morphometric_data)

print('Mutual information score:',  mutual_info_score(labels['cluster'].values, agglom.labels_))
print('Adjusted rand index: ', adjusted_rand_score(labels['cluster'].values, agglom.labels_))
```

### Clustering on density maps

```python
sns.clustermap(DM_pca[:,:n_components], method='ward', 
               row_cluster=True, 
               row_colors=[colors[t-1] for t in true_labels], 
               cbar_pos=(1., 0.6, 0.05, 0.18), 
               xticklabels=['PC%i'%i for i in range(1,n_components+1)])
```

```python
# Cluster and compare to ground truth

agglom = AgglomerativeClustering(n_clusters=14, linkage='ward')
agglom.fit(DM_pca[:,:n_components])

print('Mutual information score:',  mutual_info_score(labels['cluster'].values, agglom.labels_))
print('Adjusted rand index: ', adjusted_rand_score(labels['cluster'].values, agglom.labels_))
```

**Q:** Which representation allows for better recovery of the cell type labels? Why is that so?

**Answer:** The Z-density maps perform better than morphometrics. We can see that visually on the better ordering of the label colors in the clustermap and also quantitatively through the higher MI and adjusted Rand Index score. 
The RGC types are highly correlated with stratification profile within the IPL, hence, Z-density maps perform well.

## Predicting the projection target

```python
from sklearn.linear_model import LogisticRegressionCV


# Fit logistic regression on density maps
lr_dm = LogisticRegressionCV(scoring='balanced_accuracy')
lr_dm.fit(DM_pca, labels['projection_site'].values)
print('Average projection site prediction accuracy on density maps: %i percent'%(lr_dm.score(DM_pca, labels['projection_site'].values)*100))


# Fit logistic regression on morphometric data
lr_morph = LogisticRegressionCV(scoring='balanced_accuracy')
lr_morph.fit(morphometric_data, labels['projection_site'].values)
print('Average projection site prediction accuracy on morphometric data: %i percent'%(lr_morph.score(morphometric_data, labels['projection_site'].values)*100))
```

While Z density maps allow for better recovery of cell type labels, they are worse than morphometric statistics on predicting the projection target.

**Q:** Which morphometrics are informative on projection target?

**Answer:** All that are related to size. The cells projecting to Pbg tend to be bigger in dendritic diameter ($width \times depth$) and have wider branch angles. This is also in line with what Reinhard et al. find.

## Further references

Other ways exist to represent and compare morphologies are
* Persistence: [Description](https://link.springer.com/article/10.1007/s12021-017-9341-1) and [application on somatosensory pyramidal cell dendrites](https://academic.oup.com/cercor/article/29/4/1719/5304727) by Kanari et al. 2018

* Tree edit distance: [Heumann et al. 2009](https://link.springer.com/article/10.1007/s12021-009-9051-4)

* Sequential encoding inspired by BLAST: [Encoding](https://link.springer.com/article/10.1186/s12859-015-0604-2) and [similarity analysis on cortical dendrites](https://link.springer.com/article/10.1186/s12859-015-0605-1) by Gilette et al. 2015

* Vector point clouds: [BlastNeuron: Wan et al. 2015](https://link.springer.com/article/10.1007/s12021-015-9272-7)
